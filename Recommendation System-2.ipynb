# Content Based Filtering with Anime Dataset

```python
df_anime= pd.read_csv('anime.csv')
df_anime
```

```python
df_anime=df_anime.loc[:,['anime_id','name','rating','genre']]
```

```python
df_anime=df_anime.dropna()
```

```python
df_anime.shape
```

```python
df_anime=df_anime.reset_index(drop=True)
```

```python
cvr=CountVectorizer(tokenizer=lambda x:x.split(', '))
df_genre=cvr.fit_transform(df_anime['genre'])
df_genre=pd.DataFrame(df_genre.toarray(),columns=cvr.get_feature_names())
df_genre=pd.concat([df_anime[['anime_id','name','rating']],df_genre],axis=1)
df_genre #item feature matrix
```

```python
df_liked=df_genre[df_genre['name'].isin(['Naruto','One Piece', 'Dragon Ball'])]
df_liked
```

```python
for i in cvr.get_feature_names():
    df_liked[i]= df_liked['rating']*df_liked[i]

df_liked
```

```python
user_feature_vector=df_liked[cvr.get_feature_names()].sum()/df_liked[cvr.get_feature_names()].sum().sum()
```

```python
df_rec=df_genre.loc[~df_genre['name'].isin(['Naruto','One Piece', 'Dragon Ball'])] 
#dataframe tanpa naruto, onepiece dan dragon ball
```

```python
for i in cvr.get_feature_names():
    df_rec[i]= df_rec[i]*user_feature_vector[i]
    
df_rec
```

```python
df_rec['predicted_score']=df_rec.drop(columns=['rating','anime_id']).sum(axis=1)
```

```python
df_rec[['name','predicted_score']].sort_values('predicted_score', ascending=False).head(10)
```

```python
# as a function
def recommend_me():
    anime=input('Masukan judul anime yang anda sukai, dipisah dengan tanda koma').split(', ')
    
    import pandas as pd
    import numpy as np
    from sklearn.feature_extraction.text import CountVectorizer
    
    #buat dataframe genre
    df_anime= pd.read_csv('anime.csv')
    df_anime=df_anime.loc[:,['anime_id','name','rating','genre']].dropna().reset_index(drop=True)
    cvr=CountVectorizer(tokenizer=lambda x:x.split(', '))
    df_genre=cvr.fit_transform(df_anime['genre'])
    df_genre=pd.DataFrame(df_genre.toarray(),columns=cvr.get_feature_names())
    df_genre=pd.concat([df_anime[['anime_id','name','rating']],df_genre],axis=1)
    
    #user feature vector
    df_liked=df_genre[df_genre['name'].isin(anime)]
    
    for i in cvr.get_feature_names():
        df_liked[i]= df_liked['rating']*df_liked[i]
    
    user_feature_vector=df_liked[cvr.get_feature_names()].sum()/df_liked[cvr.get_feature_names()].sum().sum()
    
    # predict score
    
    df_rec=df_genre[~df_genre['name'].isin(anime)].drop(columns='rating')
    
    for i in cvr.get_feature_names():
        df_rec[i]= df_rec[i]*user_feature_vector[i]
        
    df_rec['predicted_score']=df_rec.drop(columns=['anime_id']).sum(axis=1)
    
    return df_rec[['name','predicted_score']].sort_values('predicted_score', ascending=False).head(10)
```

```python
recommend_me()
```

# Content Based Filtering for Multiple Users

```python
df_item_feature=pd.DataFrame({
    'movies':['Terminator 2','Interstellar','Ant Man 2','3 Idiots'],
    'action':[1,0,1,0],
    'scifi':[1,1,1,0],
    'adventure':[0,1,1,0],
    'comedy':[0,0,1,1],
    'drama':[0,1,0,1]
})
df_item_feature
```

```python
df_user_item=pd.DataFrame({
    'user':['user 1','user 2','user 3','user 4'],
    'Terminator 2':[7,8,9,0],
    'Interstellar':[9,0,0,7],
    'Ant Man 2':[8,6,0,0],
    '3 Idiots':[9,5,10,9]
})
df_user_item
```

untuk melakukan content based filtering multiple user, kita memerlukan 2 matrix: item-feature dan user-item

```python
# hitung user feature matrix
user_item=np.array(df_user_item.drop(columns='user'))
# user_item
item_feature=np.array(df_item_feature.drop(columns='movies'))
# item_feature

n_user=user_item.shape[0]
n_item= user_item.shape[1]
n_feature=item_feature.shape[1]

user_feature=np.empty((n_user,n_feature))

for i in range(n_user):
#     print(user_item[i,:])
    user_feature_vector=np.matmul(user_item[i,:], item_feature)
#     print(user_feature_vector)
    user_feature_vector=user_feature_vector/user_feature_vector.sum()
    user_feature[i,:]= user_feature_vector
```

```python
user_feature #user feature matrix
```

```python
df_user_vector=pd.DataFrame(user_feature, index=df_user_item['user'], columns=df_item_feature.columns[1:])
df_user_vector
```

```python
# prediksi score dari film yg belum pernah ditonton user
user_item_score=np.empty((n_user,n_item)) #array kosong
for i in range(n_user):
    score_=np.matmul(item_feature, user_feature[i,:])
    user_item_score[i,:]=score_

```

```python
user_item_score
```

```python
user_item_score_unwatched=np.where(user_item==0, user_item_score,0)
```

```python
pd.DataFrame(user_item_score_unwatched, columns=df_item_feature['movies'], index=df_user_item['user'])
```

Dari prediksi di atas, kita bisa merekomendasikan:
* User 3: Ant Man 2, Interstellar
* User 4: Ant Man 2 (terminator 2 terlalu rendah prediksi scorenya)
